```python

# Get the counts of each JobLevel
JobLevel_counts = data['JobLevel'].value_counts()


plt.figure(figsize=(5, 5))
plt.pie(JobLevel_counts, labels=JobLevel_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of JobLevel')
plt.axis('equal') 
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=data, x='Age', color='blue')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Age')
plt.show()
```
